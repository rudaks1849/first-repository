숫자야구 만들기 실습! (이가현, 금경만)

# 로직 정리
1. 정답 설정 (랜덤)
2. 입력 5자리 받기
3. strike / ball 판단

```python
import random
numbers = list(range(0, 10))
answer = random.sample(numbers, 5)
```

```python
print("숫자야구를 시작해보자!")
print("0부터 9까지 서로 다른 숫자 5개를 입력하세요!")
```

```python
player_numbers = list(map(int, input().strip()))
```

```python
strike = 0
ball = 0

for i in range(5):
  if player_numbers[i] == answer[i]:
    strike += 1

  elif player_numbers[i] in answer:
    ball += 1

print(answer)
#print("당신의 정답은 "+str(strike)+" strike, "+str(ball)+" ball")
print(f"당신의 정답은 {strike} strike, {ball} ball")
```
